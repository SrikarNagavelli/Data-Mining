```python
df['HRA'].fillna(df['HRA'].mean(), inplace=True)
```

```python
df['HRA'] = df['HRA'].fillna(df['HRA'].mean())
```

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python

df[['HRA', 'DA']] = scaler.fit_transform(df[['HRA', 'DA']])
```

```python
df['HRA'] = (df['HRA'] - df['HRA'].min()) / (df['HRA'].max() - df['HRA'].min())
```

```python
from sklearn.preprocessing import StandardScaler

s_scaler = StandardScaler()
```

```python
df[['TAX']] = s_scaler.fit_transform(df[[ 'TAX']])

df['TAX']
```

```python
df['HRA'] = (df['HRA'] - df['HRA'].mean()) / df['HRA'].std()
```
