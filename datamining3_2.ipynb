```python
import pandas as pd
url="/content/netflix_titles.csv"
data=pd.read_csv(url)
summary=data.describe()
print(summary)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['duration'].dropna(),edgecolor='black',bins=20)
plt.xlabel('duration')
plt.ylabel('Frequency')
plt.show()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['rating'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Netflix Class Distribution')
plt.xlabel('rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['type'].value_counts().plot(kind='bar', color='orange')
plt.title('Netflix Class Distribution')
plt.xlabel('type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['release_year'].value_counts().plot(kind='bar', color='orange')
plt.title('Netflix Class Distribution')
plt.xlabel('release_year')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['release_year'].dropna(), patch_artist=True,
boxprops=dict(facecolor='lightgreen'))
plt.title('Box Plot of year')
plt.ylabel('release')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['release_year', 'duration', 'rating']].dropna())
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['rating_encoded'] = le.fit_transform(data['rating'])
correlation_matrix = data[['release_year', 'rating_encoded']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='date_added', data=data, palette='Set2')
plt.title('date Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='type', y='release_year', data=data, palette='muted')
plt.title('type Distribution by duration year Class')
plt.show()
```
